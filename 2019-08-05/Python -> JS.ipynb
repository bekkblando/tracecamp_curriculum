```python
# Lets Loop!

loop_condition = True
loop_counter = 0
loop_end = 2

# if(loop_condition){
#   // Code
# }


while(loop_condition):
    if(loop_counter >= loop_end):
        print("\n\n Loop Condition Set to false \n\n")
        loop_condition = False
    else:
        print("Continue")
    print("Current value of loop counter", loop_counter)
    loop_counter += 1


# We'll we don't need the if statement
loop_counter = 0
while(loop_counter <= loop_end):
    print("Going through loop 2")
    loop_counter += 1
    
# loop_counter += 1

    
```

```python
# We can loop and make a list at the same time
reverse_me = "ofni ezinagro"

output = [character for character in reverse_me[::-1] if character == "o"]

new_test_list = []
for character in reverse_me[::-1]:
    if(character == 'o'):
        new_test_list.append(character)
    elif(character == 'n'):
        new_test_list.append("We just encountered an n")
    else:
        print("Any letter other than 'o' or 'n'")

print("New Test List", new_test_list)

print(reverse_me[::-1])

# print("First Char", reverse_me[-1:0])

print("OUTPUT: ", output)

print("JOINBY".join(output))
print("J,O,I,N,B,Y".split(","))
```
